Prediction using Decision Tree Algorithm

## Objective: To classify any new data correctly to its label, according to its feature values. 



```python
#importing required libraries
import sklearn.datasets as datasets
import pandas as pd
```

```python
# Loading the iris dataset
iris=datasets.load_iris()
iris
```

## Data description:
### Target Names: (['setosa', 'versicolor', 'virginica'])
### setosa:0
### versicolor:1
### virginica:2

## Forming the iris dataframe:

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names) #storing feature values in a dataframe, df.
print(df.head())

y=iris.target #storing target values in a variable, y.
print(y)
```

## Defining the decision tree algorithm 

```python
from sklearn.tree import DecisionTreeClassifier
dtree_cls=DecisionTreeClassifier()
dtree_cls.fit(df,y)
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree_cls, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### The degree of Gini index varies between 0 and 1, where 0 denotes that all elements belong to a certain class or if there exists only one class, and 1 denotes that the elements are randomly distributed across various classes. A Gini Index of 0.5 denotes equally distributed elements into some classes.

```python
prob=dtree_cls.predict_proba([[6,3,4.5,1.6]]) #[[sepal length(cm),sepal width(cm),petal length(cm),petal width(cm)]]
print(prob)
```

```python
#Inference: The probability is 1 for second label in the array of target names, ie versicolor.
#while probability is zero for rest of the two.
#Hence, the Decision Tree Classifier is able to classify the given feature values with 100% accuracy.
```
